```python
import json
from collections import OrderedDict
```

```python
with open('./data.json', encoding='utf-8') as f:
    dict_plain = json.loads(f.read(), object_pairs_hook=OrderedDict)
```

```python
dict_plain
```

```python
dict_plain_try = OrderedDict([('110000', '北京市'),
             ('110101', '东城区'),
             ('110102', '西城区'),
             ('110105', '朝阳区'),
             ('110106', '丰台区'),
             ('110107', '石景山区'),
             ('110108', '海淀区'),
             ('110109', '门头沟区'),
             ('110111', '房山区'),
             ('110112', '通州区'),
             ('110113', '顺义区'),
             ('110114', '昌平区'),
             ('110115', '大兴区'),
             ('110116', '怀柔区'),
             ('110117', '平谷区'),
             ('110118', '密云区'),
             ('110119', '延庆区'),
             ('120000', '天津市'),
             ('120101', '和平区'),
             ('120102', '河东区'),
             ('120103', '河西区')])
```

```python
res = {}
sheng_ming = ''
shi_ming = ''
xian_ming = ''
sheng_xu = -1
shi_xu = -1
xian_xu = -1
sheng_template = {'id':'','name':'','shi':{}}
shi_template = {'id':'','name':'','xian':{}}
xian_template = {'id':'','name':''}
for code,name in dict_plain_try.items():
    sheng_code = code[0:2]
    shi_code = code[2:4]
    xian_code = code[4:]
    if shi_code == '00' and xian_code == '00':
        # initialize sheng
        sheng_xu += 1
        res.update({sheng_xu: sheng_template})
        res[sheng_xu]['id'] = sheng_code
        res[sheng_xu]['name'] = sheng_ming
        res[sheng_xu]['shi'].update({0: shi_template})
    elif xian_code == '00':
        shi_xu += 1
        shi_ming = name
        res[sheng_xu]['shi'].update({shi_xu: {'id': shi_code, 'name': shi_ming, 'xian': xian_template}})
    else:
        if shi_xu == -1: shi_xu = 0
        xian_xu += 1
        xian_ming = name
        res[sheng_xu]['shi'][shi_xu]['xian'].update({xian_xu: {'id': xian_code, 'name': xian_ming}})     
```

```python
res
```

```python
with open('./data_structured.json', encoding='utf-8', mode='w') as g:
    json.dump(res, g, indent=4, ensure_ascii=False)
```
